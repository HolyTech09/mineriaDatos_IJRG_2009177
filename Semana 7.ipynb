```python
import pandas as pd
from statsmodels.formula.api import ols
import statsmodels.api as sm

df = pd.read_csv("csv/aprovCreditos.csv", index_col=False)

colNumericas = [ "Approved_Declined_Amount", "Disbursed_Shipped_Amount"]

colCategoricas = ["Program", "Country", "Term"]

        
print("Imprimiendo anova")
for cn in colNumericas:
    for cc in colCategoricas:
        modl = ols(f"{cn} ~ {cc}", data=df).fit()
        anova_df = sm.stats.anova_lm(modl, typ=2)
        print(f"Entre {cc} y {cn}")
        if anova_df["PR(>F)"][cc]< 0.005:
            print("Hay diferencias")
            print(anova_df)
        else:
            print("No hay diferencias")
        print()
print("Imprimiendo T test")
for cc in colCategoricas:
    for cn in colNumericas:
        topCat = df.groupby(cc)[cn].mean().sort_values(ascending = False).head(2).index
        group1 = df[df[cc]==topCat[0]][cn]
        group2 = df[df[cc]==topCat[1]][cn]
        t_stat, p_value, df_freedom = sm.stats.ttest_ind(group1, group2)
        if p_value < 0.005:
            print(f"Hay diferencias entre los grupos {topCat[0]} y {topCat[1]}")
        else:
            print(f"No existen diferencias entre los grupos {topCat[0]} y {topCat[1]}")
        print()
```
