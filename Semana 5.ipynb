```python
import pandas as pd
import os 
import matplotlib.pyplot as plt
from tabulate import tabulate

def print_tabulate(df:pd.DataFrame)-> pd.DataFrame:
    print(tabulate(df, headers= df.columns, tablefmt='orgtbl'))

df = pd.read_csv("csv/aprovCreditos.csv", index_col=False)

path = "img/Semana 5"
os.makedirs(path, exist_ok=True)
print_tabulate(df.describe())

colNumericas = [ "Approved-Declined Amount", "Disbursed-Shipped Amount"]

colCategoricas = list()
for col in df.columns:
    if col not in colNumericas:
        colCategoricas.append(col)

dfbyProgram = df.groupby("Program").agg({
    "Approved_Declined_Amount": ["sum", "count", "mean", "min", "max"]
})

dfByCountry = df.groupby("Country").agg({
    "Approved_Declined_Amount": ["sum", "count", "mean", "min", "max"]
})
print("Organizando por programa")
print_tabulate(dfbyProgram)
print("Organizando por pais")
print_tabulate(dfByCountry)

df.boxplot(by="Program", column=["Approved_Declined_Amount"], figsize=(18,9))
plt.xticks(rotation=90)
plt.savefig(f"{path}/boxplotProgram.png")
plt.close()
df.boxplot(by="Deal_Cancelled", column=["Approved_Declined_Amount"], figsize=(18,9))
plt.xticks(rotation=90)
plt.savefig(f"{path}/boxplotCountry.png")
plt.close()

```
