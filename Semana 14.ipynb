```python
from wordcloud import WordCloud
from collections import Counter
import matplotlib.pyplot as plt
import pandas as pd
import os

path = "img/Semana 14"
def generate_word_cloud(df, columna, filename):
    all_words = ""
    palabras = df[columna].dropna().astype(str).tolist()
    for arg in palabras:
        tokens = arg.split()
        all_words += " ".join(tokens) + " "

    print(f"\nTop 10 palabras en '{columna}':")
    print(Counter(all_words.split()).most_common(10))

    wordcloud = WordCloud(
        background_color="white", min_font_size=5
    ).generate(all_words)

    plt.close()
    plt.figure(figsize=(5, 5), facecolor=None)
    plt.imshow(wordcloud)
    plt.axis("off")
    plt.tight_layout(pad=0)
    plt.savefig(f"{path}/{filename}.png")
    plt.close()

df = pd.read_csv("csv/aprovCreditos.csv", index_col=False)
os.makedirs(path, exist_ok=True)

columnas_numeros = df.select_dtypes(include="number").columns.to_list()
columnas_texto = df.drop(columns=columnas_numeros).columns.to_list()
for col in columnas_texto:
    generate_word_cloud(df, col, f"wordcloud_{col.replace(' ', '_')}")
```
